<h4>Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot. 

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.title("This is a scatter plot")
plt.xlabel("This is x-axis")
plt.ylabel("This is y-axis")
```

<h4>Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

Ans. The subplot() function in Matplotlib is a versatile function used to create multiple plots in a single figure. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# For line 1: 
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)

# For line 2: 
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)

# For line 3:
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)

# For line 4:
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)
```

<h4>Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

Ans. A bar plot or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.A bar graph shows comparisons among discrete categories. One axis of the chart shows the specific categories being compared, and the other axis represents a measured value.

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(4,2))
colors=['r','g','b','y']
plt.bar(company,profit,color=colors)
plt.xlabel("Name of company")
plt.ylabel("Profit earned")
plt.title("Bar Graph")
```

```python
plt.figure(figsize=(5,3))
colors=['r','g','b','y']
plt.barh(company,profit,color=colors)
plt.ylabel("Name of company")
plt.xlabel("Profit earned")
plt.title("Horizontal Bar Graph")
```
